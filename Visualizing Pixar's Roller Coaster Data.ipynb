# Data Cleaning

For now, lets do some data cleaning, I have noticed that the last row, describing the movie Inside Out, is missing values for Oscars Nominated and Oscars Won (NaN values), I will avoid using Inside out. Also the IMDB Score column is on a 10 point scale while the RT Score and Metacritic Score columns are on a 100 point scale. We need to convert the IMDB scores to a 100 point scale so we can compare how the movies fared across these different review sites. 

Let's clean up the pixar_movies DataFrame first then create the filtered DataFrame containing the first 14 rows at the end, so we don't have to repeat the cleanup for both DataFrames.

```python
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies["International %"] = pixar_movies["International %"].str.rstrip("%").astype("float")
```

Now we Multiply the `IMDB Score` column by 10

```python
pixar_movies["IMDB Score"] = pixar_movies["IMDB Score"]*10
```

We create a new DataFrame containing only the first 14 rows, and we assing this new dataframe to filtered_pixar.

```python
filtered_pixar = pixar_movies.loc[0:13]
```

We have to set the `Movie` column as the index for the DataFrame.

```python
pixar_movies.set_index("Movie", inplace=True)
filtered_pixar.set_index("Movie", inplace=True)
```

We check if the data cleaning was succesful by showing the new dataframe.

```python
filtered_pixar
```

# Data visualization, line plots

Now that the average reviews for each movie from the 3 major review sites are all on a 100 point scale, we can ask the 
question: How do the Pixar films fare across each of the major review sites? Let's generate the following 3 line plots:
Rotten Tomatoes score on the y-axis, movie name on the x-axis, IMDB score on the y-axis, movie name on the x-axis, Metacritic score on the y-axis, movie name on the x-axis.

Let's plot these line plots on the same plotting area so we can easily compare how each movie fared across the review sites.

```python
critics_reviews = pixar_movies[["RT Score","Metacritic Score","IMDB Score"]]
critics_reviews.plot()
```

The resulting chart is a little cramped, so let's tweak the figsize parameter and increase the size. When calling the plot() method, we set the figsize parameter to (6,10) so the plot is displayed across 6 inches wide and 10 inches tall.
Here's how this bigger plot looks like:

```python
critics_reviews.plot(figsize=(10,6))
```

# Data visualization, box plot

From the previous plot, it seems like the review site Rotten Tomatoes gives Pixar consistently higher ratings. Let's generate a box plot to explore the question: How are the average ratings from each review site across all the movies distributed? A box plot is useful for us here since we can visually compare the box part to determine where the central tendency of each distribution lies and the whiskers to see how spread out each distribution is.

```python
pixar_movies[["RT Score","Metacritic Score","IMDB Score"]].plot(kind="box")
```

If we set the figsize parameter to (9,5) we will generate a nicer looking plot. Here's how the new box plot looks like:

```python
pixar_movies[["RT Score","Metacritic Score","IMDB Score"]].plot(kind="box", figsize=(9,5))
```

From the box plot, it looks like our hunch that Rotten Tomatoes gave the Pixar movie consistently higher ratings seems to be true. It also looks like Metacritic's ratings for the Pixar movies are more spread out, ranging from right under 60 to right under 100.

The Domestic % and International % columns describe the percentage of the total revenue that was made in the United States and internationally, respectively. We can use those columns to explore these questions: How has the ratio of where the revenue comes from changed since the first movie? Now that Pixar is more well known internationally, is more revenue being made internationally for newer movies?

# Data visualization, stacked bar plot

```python
revenue_proportions = filtered_pixar[["Domestic %", "International %"]]
revenue_proportions.plot(kind='bar', stacked=True, figsize=(7,6))
```

When we spend some time explorering the stacked bar plot we notice that there's been a general decrease in the proportion of revenue that was made domestically. While the first few movies were right around 50%, the last few movies have been hovering closer to 40%.
